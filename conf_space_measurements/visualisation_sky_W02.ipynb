```python
import numpy as np
import matplotlib.pyplot as plt
import time
import healpy as hp
from astropy.table import Table
from astropy.io import fits
```

```python
hdu_list = fits.open('../../fits/sky_W_W02.fits', memmap=True)
hdu_list.info()
evt_data = Table(hdu_list[1].data)
ra = evt_data['ra']
dec = evt_data['dec']
```

```python
plt.plot(ra,dec,'.')
plt.show()
```

```python
# computing area
def get_area(nside):
    npix = hp.nside2npix(nside)
    theta = np.radians(90. - dec)
    phi = np.radians(ra)
    ipix = hp.ang2pix(nside, theta, phi)
    area = ipix.size*hp.nside2pixarea(nside,degrees=True)
    print(area, ipix.size,npix)
```

```python
get_area(256)
get_area(512)
get_area(1024)
get_area(2048)
```
