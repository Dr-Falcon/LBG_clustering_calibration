```python
import numpy as np
import matplotlib.pyplot as plt
import time
import healpy as hp
from astropy.table import Table
from astropy.io import fits
```

```python
#get data vector
data = {}

def get_SDSS_seeing(path):
    
    #hdu_list = fits.open('../fits/sdss_seeing_psf.fits', memmap=True)
    hdu_list = fits.open(path, memmap=True)
    
    #hdu_list.info()
    evt_data = Table(hdu_list[1].data)
    
    ra = evt_data['ra']
    dec = evt_data['dec']    
    shape11=evt_data['i_sdssshape_psf_shape11']
    shape22=evt_data['i_sdssshape_psf_shape22']
    shape12=evt_data['i_sdssshape_psf_shape12']
    
    det_rad=(shape11*shape22-shape12**2)**0.25
    FWHM = 2*np.sqrt(2.*np.log(2))*det_rad
    
    #removing nan
    FWHM = FWHM[np.isnan(FWHM)==False]
    
    data['seeing'] = {'ra':ra, 'dec':dec, 'FWHM':FWHM}

def get_data(path):
    with open(path) as f:
    line = [s.strip() for s in f.readlines()]

    f = open(path,'r')
    nheader = 167
    
    ra_LBGs = []
    dec_LBGs = []
    i_cmodel_mag_LBGs = []
    
    ra_stars = []
    dec_stars = []
    
    n_LBGs = 0
    n_stars = 0
    
    for j in range(len(line)):
            list = f.readline()
            columns = list.split()
            if j>=nheader:
                
                ra = columns[1]
                dec = columns[2]
                i_cmodel = column[27]
                i_convolvedflux_0_20_mag = float(columns[149])
                a_i = float(columns[87])

                #extendedness
                i_extendedness = columns[121]
                
                # flux cut for data
                if i_convolvedflux_0_20_mag - a_i < 25.5:
                    n_LBGs += 1
                    ra_LBGs.append(ra)
                    dec_LBGs.append(dec)
                    i_cmodel_mag_LBGs.append(i_cmodel)
                    
                    if i_extendedness == '0.0':
                        n_stars += 1
                        ra_stars.append(ra)
                        dec_stars.append(dec)
            
        #ndarray conversion
    ra_LBGs = np.array(ra_LBGs)
    dec_LBGs = dec.array(dec_LBGs)
    ra_stars = np.array(ra_stars)
    dec_stars = np.array(dec_stars)
    i_cmodel_mag_LBGs = np.array(i_cmodel_mag_LBGs)
    
    # average density of LBGs
    area = 
    n_ave = n_LBGs/area
    
    # labeling
    data['LBGs'] = {'ra':ra_LBGs, 'dec':dec_LBGs, 'i_cmodel': i_cmodel_mag_LBGs, 'n_ave': n_ave}
    data['stars'] = {'ra':ra_stars, 'dec':dec_stars}


# execute functions to generate a data vector
get_SDSS_seeing('../../fits/sdss_seeing_psf.fits')
get_data('../../data/goldrush/SigmaClippedAround_SNcutted_gri_W_W02.cat')
```
