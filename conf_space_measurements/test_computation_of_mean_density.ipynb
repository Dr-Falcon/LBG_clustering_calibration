```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
```

```python
# Generate random data (RA, Dec) in degrees
np.random.seed(42)
n_points = 1000
ra = np.random.uniform(0, 10, n_points)
dec = np.random.uniform(-5, 5, n_points)
```

```python
plt.plot(ra,dec,'.')
plt.show()
```

```python
def create_map(ra,dec,nside):
    # Healpix parameters
    print(nside)
    npix = hp.nside2npix(nside)
    theta_pix = np.sqrt(hp.nside2pixarea(nside,degrees=True))
    #print(theta_pix)
    # Convert (theta, phi) to pixel indices
    theta = np.radians(90 - dec)
    phi = np.radians(ra)
    ipix = hp.ang2pix(nside, theta, phi)
    #map
    map = np.zeros(npix)
    for i in ipix:
        print(i)
        map[i]+=1
    
    return map
```

```python
create_map(ra,dec,256)
create_map(ra,dec,512)
create_map(ra,dec,1024)
```
