```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vs_nstars(band,nside):
    mean, var, nstars = np.loadtxt('./nLBGs_per_n_rand-nstar_%s-band_nside%d_all-areas.dat'%(band,nside), delimiter=' ',unpack=True)
    std = np.sqrt(var)
    #plt.title(r'$%s-band n_{\rm LBGs}/n_{\rm random}$ vs. $n_{\rm stars}$ for $n_{\rm side} = %d$ at %s'%(band, nside, sky))
    plt.xlabel(r'$n_{\rm stars} {\rm [/deg^2]}$')
    plt.ylabel(r'$n_{\rm LBGs}/n_{\rm random}$')
    #plt.xlim([0.,1000.])
    plt.ylim([-1.,3])
    plt.errorbar(nstars, mean, yerr=std, fmt='.')
    #plt.xscale('log')
    plt.legend()
    #plt.show()
    
def plot_vs_extendedness0(band,nside):
    mean, var, nstars = np.loadtxt('./nLBGs_per_n_rand-extendedness0_%s-band_nside%d__all-areas.dat'%(band,nside), delimiter=' ',unpack=True)
    std = np.sqrt(var)
    #plt.title(r'$%s-band n_{\rm LBGs}/n_{\rm random}$ vs. $n_{\rm stars}$ for $n_{\rm side} = %d$ at %s'%(band, nside, sky))
    plt.xlabel(r'$n_{\rm stars} {\rm [/deg^2]}$')
    plt.ylabel(r'$n_{\rm LBGs}/n_{\rm random}$')
    #plt.xlim([0.,1000.])
    plt.ylim([-1.,3])
    plt.errorbar(nstars, mean, yerr=std, fmt='.')
    #plt.xscale('log')
    plt.legend()
    #plt.show()
    
def plot_vs_fluxsigma2(band,nside):
    mean, var, sigma2 = np.loadtxt('./nLBGs_per_n_rand-fluxsigma2_%s-band_nside%d_all-areas.dat'%(band, nside), delimiter=' ',unpack=True)
    std = np.sqrt(var)
    #plt.title(r'%s-band $n_{\rm LBGs}/n_{\rm random}$ vs. $<\sigma^2>$ for $n_{\rm side} = %d$ at %s'%(band, nside, sky))
    plt.xlabel(r'$<\sigma^2>$')
    plt.ylabel(r'$n_{\rm LBGs}/n_{\rm random}$')
    #plt.xlim([0.,1000.])
    #plt.ylim([-1.,3])
    plt.errorbar(sigma2, mean, yerr=std, fmt='.')
    plt.xscale('log')
    plt.legend()
    #plt.show()
    
def plot_vs_i_sdss_seeing(band,nside):
    mean, var, sigma2 = np.loadtxt('./nLBGs_per_n_rand-sdss_seeing_%s-band_nside%d_all-areas.dat'%(band, nside), delimiter=' ',unpack=True)
    std = np.sqrt(var)
    plt.title(r'%s-band $n_{\rm LBGs}/n_{\rm random}$ vs. FWHM for $n_{\rm side} = %d$'%(band, nside))
    plt.xlabel(r'FWHM')
    plt.ylabel(r'$n_{\rm LBGs}/n_{\rm random}$')
    #plt.xlim([0.,1000.])
    #plt.ylim([-1.,3])
    plt.errorbar(sigma2, mean, yerr=std, fmt='.')
    plt.show()
```
